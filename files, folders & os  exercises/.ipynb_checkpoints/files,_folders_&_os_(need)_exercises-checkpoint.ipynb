Task 1

```python
import os
import glob
import shutil
```

```python
month_mapping = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
}
path = os.path.join('.', 'os-collaboration-exercise-data')
for filename in os.listdir('os-collaboration-exercise-data'):
    parts = filename.split('-')
    if len(parts) == 2:
        month, date_city = parts
        date, city = date_city.split('_')
        if month.lower() in month_mapping:
            month_numeric = month_mapping[month.lower()]
            new_filename = f"{month_numeric}-{date}-{city}"
            new_filepath = os.path.join('.', 'os-collaboration-exercise-data', new_filename)
            os.rename(os.path.join('.', 'os-collaboration-exercise-data', filename), new_filepath)
```

Task 2

```python
def rename_my_file(old_file_name):
    month_mapping = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
    }
    path = os.path.join('.', 'os-collaboration-exercise-data')
    for filename in os.listdir('os-collaboration-exercise-data'):
        parts = old_file_name.split('-')
        if len(parts) == 2:
            month, date_city = parts
            date, city = date_city.split('_')
            if month.lower() in month_mapping:
                month_numeric = month_mapping[month.lower()]
                new_filename = f"{month_numeric}-{date}-{city}"
                return new_filename
```

Task 3

```python
lst = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']

for old_file_name in lst:
    new_file_name = rename_my_file(old_file_name)
    print(f"{old_file_name} ----> {new_file_name}")
```

Task 4

```python
path = os.path.join('.', 'os-collaboration-exercise-data')
txt_files = glob.glob(os.path.join(path, '*.txt'))
try:
    for old_file_path in txt_files:
        old_file_name = os.path.basename(old_file_path)
        new_file_name = rename_my_file(old_file_name)
except:
    print('The files have been renamed already')
```

```python
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for city in cities:
    city_folder = os.path.join('os-collaboration-exercise-data', city)
    print(f'Creating {city_folder}')
    os.mkdir(city_folder)
```

Task 6

```python
cairo_files = glob.glob('os-collaboration-exercise-data/*cairo*', recursive=True)
print(cairo_files)
```

```python
for cairo in cairo_files:
    path_to_destination = os.path.join('.','os-collaboration-exercise-data','Cairo')
    shutil.copy(cairo, path_to_destination)
    print(f'Copied file to {path_to_destination}')
```
